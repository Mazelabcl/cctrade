```python
import pandas as pd
```

```python
levels_2 = 'levels_dataset_2017_01_01-2020_12_31.csv'
levels_1 = 'levels_dataset_2021_01_01-2024_12_31.csv'

ml_2 = 'ml_dataset_2017_01_01-2020_12_31.csv'
ml_1 = 'ml_dataset_2021_01_01-2024_12_31.csv'
```

```python
# Read the CSV files into DataFrames
levels_1_df = pd.read_csv(levels_1)
levels_2_df = pd.read_csv(levels_2)

# Concatenate the DataFrames
levels_dataset = pd.concat([levels_1_df, levels_2_df], axis=0)

# Reset the index
levels_dataset.reset_index(drop=True, inplace=True)
```

```python
# Display the concatenated DataFrame
levels_dataset.head()
```

```python
levels_dataset.to_csv('levels_dataset.csv', index=False)
```

```python
# Read the CSV files into DataFrames
ml_1_df = pd.read_csv(ml_1)
ml_2_df = pd.read_csv(ml_2)

# Concatenate the DataFrames
ml_dataset = pd.concat([ml_1_df, ml_2_df], axis=0)

# Reset the index
ml_dataset.reset_index(drop=True, inplace=True)
```

```python
# Display the concatenated DataFrame
ml_dataset.head()
```

```python
ml_dataset.to_csv('ml_dataset.csv', index=False)
```
